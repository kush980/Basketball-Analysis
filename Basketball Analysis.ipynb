```python
def myplot(data , Playerlist = Players):
    cdic={'KobeBryant':'green',"JoeJohnson":'red',"LeBronJames":'magenta',"CarmeloAnthony":'black',"DwightHoward":'yellow',"ChrisBosh":'grey',"ChrisPaul":'orange',"KevinDurant":'silver',"DerrickRose":'cyan',"DwayneWade":'blue'}
    mark={"KobeBryant":'s',"JoeJohnson":'o',"LeBronJames":'^',"CarmeloAnthony":'*',"DwightHoward":'h',"ChrisBosh":'P',"ChrisPaul":'p',"KevinDurant":'X',"DerrickRose":'o',"DwayneWade":'*'}
    for name in Playerlist:
        plt.plot(data[Pdict[name]] ,c=cdic[name],ls='--',marker=mark[name],ms=7,label=Players[Pdict[name]])
    plt.legend(loc='upper left',bbox_to_anchor=(1,1))
    plt.xlabel('Year')
    plt.ylabel('data')
    plt.title('Yearwise comparision of Basketball Players')
    plt.xticks(list(range(0,10)),Seasons,rotation='horizontal')
    plt.show()
```
